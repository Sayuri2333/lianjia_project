```python
import requests
import requests
import time
from lxml import etree
```

```python
#设置列表页URL的固定部分
url='http://gz.lianjia.com/chengjiao/'
#设置页面页的可变部分
page=('pg')
#设置请求头部信息
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
'Accept':'text/html;q=0.9,*/*;q=0.8',
'Accept-Charset':'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
'Accept-Encoding':'gzip',
'Connection':'close',
'Referer':'http://www.baidu.com/link?url=_andhfsjjjKRgEWkj7i9cFmYYGsisrnm2A-TN3XZDQXxvGsM9k9ZZSnikW2Yds4s&amp;amp;wd=&amp;amp;eqid=c3435a7d00146bd600000003582bfd1f'
}
for i in range(1,80):
    if i == 1:
        i=str(i)
        a=(url+page+i+'/')
        r=requests.get(url=a,headers=headers)
        html=r.content
    else:
        i=str(i)
        a=(url+page+i+'/')
        r=requests.get(url=a,headers=headers)
        html2=r.content
        html = html + html2
    #每次间隔0.5秒
    time.sleep(0.5)
encoding=requests.get(url,headers=headers).encoding
lj=etree.HTML(html,parser=etree.HTMLParser(encoding=encoding))
#提取房源总价
price=lj.xpath('//div[@class="totalPrice"]')

tp=[]
for a in price:
    totalPrice=a.xpath('.//span/text()')[0]
    tp.append(totalPrice)
for p in tp[:10]:
    print(p)
```

```python
#提取房源单价
price=lj.xpath('//div[@class="unitPrice"]')

up=[]
for a in price:
    totalPrice=a.xpath('.//span/text()')[0]
    up.append(totalPrice)
for p in up[:10]:
    print(p)
```

```python
#提取房源信息
houseInfo=lj.xpath('//div[@class="houseInfo"]')

hi=[]
for b in houseInfo:
    house=b.xpath('.//text()')[0]
    hi.append(house)
for i in hi[:10]:
    print(i)
```

```python
# 提取位置信息
positionInfo=lj.xpath('//div[@class="positionInfo"]')

pi=[]
for b in positionInfo:
    house=b.xpath('.//text()')[0]
    pi.append(house)
for i in pi[:10]:
    print(i)
```

```python
#导入pandas库
import pandas as pd
#创建数据表
house=pd.DataFrame({'unitprice':up,'totalprice':tp,'houseinfo':hi,'positioninfo':pi})
#查看数据表的内容
house.head()
```

```python
#对房源信息进行分列
houseinfo_split = pd.DataFrame((x.split('|') for x in house.houseinfo),index=house.index,columns=['chaoxiang','zhuangxiu','dianti'])
houseinfo_split.head()
```

```python
#将分列结果拼接回原数据表
house=pd.merge(house,houseinfo_split,right_index=True, left_index=True)
house.head(20)
```

```python
#对位置信息进行分列
import re
positioninfo_split = pd.DataFrame((re.split(r"楼.+\)", x, maxsplit=0)[0] for x in house.positioninfo),index=house.index,columns=['louceng'])
positioninfo_split.head()
```

```python
house=pd.merge(house,positioninfo_split,right_index=True, left_index=True)
house.head(20)
```

```python
# 提取房名信息
housingInfo=lj.xpath('//div[@class="title"]')

hoi=[]
for b in housingInfo:
    hou=b.xpath('.//a/text()')
    hoi.append(hou)
hoi = [x[0] for x in hoi if x!=[]]
hoi = [x for x in hoi if x!='去登录']
for i in hoi[:10]:
    print(i)
```

```python
housingInfo=pd.DataFrame({'allinfo':hoi})
#查看数据表的内容
housingInfo.head()
```

```python
housinginfo_split = pd.DataFrame((x.split(' ') for x in housingInfo.allinfo),index=housingInfo.index,columns=['xiaoqu','huxing','daxiao'])
housinginfo_split.head()
```

```python
daxiao_split = pd.DataFrame((x.split('平米') for x in housinginfo_split.daxiao),index=housingInfo.index,columns=['size','a'])
daxiao_split.drop(['a'],axis=1,inplace=True)
daxiao_split.head()
```

```python
huxing_split = pd.DataFrame((re.split('["室","厅"]',x) for x in housinginfo_split.huxing),index=housingInfo.index,columns=['main_room','sub_room', 'a'])
huxing_split.drop(['a'],axis=1,inplace=True)
huxing_split.head()
```

```python
housinginfo_split.drop(['huxing','daxiao'],axis=1,inplace=True)
housinginfo_split.head()
```

```python
house=pd.merge(house,housinginfo_split,right_index=True, left_index=True)
house.head()
```

```python
house=pd.merge(house,daxiao_split,right_index=True, left_index=True)
house.head()
```

```python
house=pd.merge(house,huxing_split,right_index=True, left_index=True)
house.head()
```

```python
house['unitprice']=pd.to_numeric(house['unitprice'],errors='coerce')
```

```python
house.dtypes
```

```python
house['totalprice']=house['totalprice'].astype('float64')
house.dtypes
```

```python
house['size']=pd.to_numeric(house['size'],errors='coerce')
house['main_room']=pd.to_numeric(house['main_room'],errors='coerce')
house['sub_room']=pd.to_numeric(house['sub_room'],errors='coerce')
```

```python
house.dtypes
```

```python
house.to_csv("house.csv",encoding ="utf-8")
```
