```python
import pandas as pd
house_baoan = pd.read_excel('house_baoan.xls', index_col=None)
house_baoan.head()
```

```python
house_futian = pd.read_excel('house_futian.xls', index_col=None)
house_futian.head()
```

```python
house_longgang = pd.read_excel('house_longgang.xls', index_col=None)
house_longgang.head()
```

```python
house_longhua = pd.read_excel('house_longhua.xls', index_col=None)
house_longhua.head()
```

```python
house_luohu = pd.read_excel('house_luohu.xls', index_col=None)
house_luohu.head()
```

```python
house_nanshan = pd.read_excel('house_nanshan.xls', index_col=None)
house_nanshan.head()
```

```python
house_yantian = pd.read_excel('house_yantian.xls', index_col=None)
house_yantian.head()
```

```python
house_baoan['city'] = '宝安'
house_futian['city'] = '福田'
house_longgang['city'] = '龙岗'
house_longhua['city'] = '龙华'
house_luohu['city'] = '罗湖'
house_nanshan['city'] = '南山'
house_yantian['city'] = '盐田'
```

```python
house = pd.concat([house_baoan,house_futian,house_longgang,house_longhua,house_luohu,house_nanshan,house_yantian],axis=0,ignore_index=True)
house.head()
```

```python
house.to_excel('house.xls',sheet_name='biubiu')
```
